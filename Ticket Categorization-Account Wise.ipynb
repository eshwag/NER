```python
from NER import CRF
import pandas as pd
import time
from sklearn.model_selection import train_test_split
```

```python
%%time
consol_data = pd.read_excel('Data/Consolidated Details_GSD TDA Accounts.xlsx')
```

```python
consol_data = consol_data[['Short Description', 'Product', 'Issues', 'Account']]
consol_data.columns = ['short_description', 'Products','Issues', 'Account']
```

```python
a = CRF.ConditionalRandomField(num_features=2,sent_translate='No')
```

```python
consol_data.Account.value_counts()
```

```python
account = 'Uniper'
account_data = consol_data[consol_data['Account'] == account]
account_data = account_data[['short_description', 'Products','Issues']]
data, test = train_test_split(account_data, test_size=0.3, random_state=42)
print(f'Train Report for Account {account}')
a.train(data=data.short_description.astype('str'),
        y = [data.Products.astype('str'),data.Issues.astype('str')],
        tag = ['P','I'])
print(f'Test Report for Account {account}')
a.predict(data=test.short_description.astype('str'),
          y = [test.Products.astype('str'),test.Issues.astype('str')],
          tag = ['P','I'])
    
```
